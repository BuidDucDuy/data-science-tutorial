```python
import pandas as pd 
import numpy as np
```

```python
df  = pd.read_csv("chipotle.tsv" , sep = "\t")
```

```python
df.head(5)
```

```python
df.loc[(df.quantity == 15) | (df.item_name == "Izze")]
```

```python
df.order_id = 1
```

```python
df.iloc[5]
```

```python
#data Manipulation
df.item_price.dtype
```

```python
#lambda
df.item_price =  df.item_price.apply(lambda x : float(x.replace("$" , "")))
```

```python
df.head(4)
```

```python
df["total_price"] = df["quantity"] * df["item_price"] 
```

```python
df.head(5)
```

```python
# how much total revenue in the dataset
revenue = df["total_price"].sum()
print(revenue)
```

```python
#which was order the most item 
# user "Group By"
Order_the_most_item = df.groupby("item_name")["quantity"].sum()
Order_the_most_item.sort_values(ascending = False).head(5)
```
